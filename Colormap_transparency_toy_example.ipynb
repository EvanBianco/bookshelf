```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# get colormap
ncolors = 256
color_array = plt.get_cmap('RdBu')(range(ncolors))

# change alpha values
color_array[:,-1] = np.linspace(0.0,1.0,ncolors)

# create a colormap object
map_object = LinearSegmentedColormap.from_list(name='RdBu_alpha', colors=color_array)

# register this new colormap with matplotlib
plt.register_cmap(cmap=map_object)

def f(x, y):
    """This is some function we want to plot...it could be any thing you
    want to add a colormap to just as an amplitude extraction or a horizon
    elevation map"""
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)

# This has a transparency ramp increasing from high (opaque) to low (full transparency) values
fig, ax = plt.subplots()
h = ax.imshow(f(X,Y),cmap='RdBu_alpha')
plt.colorbar(mappable=h)

# Notice how we can't see through the blues (highs), but we can see through 
# the reds (lows)
underneath_map = np.random.rand(X.shape[0], X.shape[1])
fig, ax = plt.subplots()
ax.imshow(underneath_map, cmap='Greys')
h = ax.imshow(f(X,Y),cmap='RdBu_alpha')
plt.colorbar(mappable=h)
```
